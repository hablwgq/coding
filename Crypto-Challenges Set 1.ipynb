<i><b>Excercise 1:</b> Convert hex to Base64</i>

```python
import codecs
def hextobase64(inp):
    return codecs.encode(codecs.decode(inp, 'hex'), 'base64').decode()
```

<i><b>Excercise 4:</b> Detect single-character XOR</i>  

```python
with open("excercise4_data.txt", "r") as f:
    for i, line in enumerate(f):
        line = binascii.unhexlify(line[:-1]) # a hack to not use the "LN" character
        res1 = {}
        for char in range(256):
            res1[char] = score(singlexor(line, char))
        sortres1 = [(k, res1[k]) for k in sorted(res1, key=res1.get, reverse=True)]
        print("Line {:03d}: Key {}: {}".format(i+1,sortres1[0][0], singlexor(line, sortres1[0][0])))
```

<i><b>Excercise 5:</b> Implement repeating-key XOR</i>

```python
def repxor(inp, key):
    res = ""
    for i, char in enumerate(inp):
        res += str(codecs.decode(chr(codecs.encode(codecs.encode(key[i%len(key)]), "hex")[0] ^ codecs.encode(codecs.encode(char), "hex")[0])))
    return res
```
